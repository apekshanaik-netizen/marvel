## Quest 2: Python Coding & Visualization

###  Section Overview

In this section, I will perform basic Python operations such as variable declaration and arithmetic calculations. I will also create a simple line graph using Matplotlib to visualize data.

This demonstrates how Jupyter Notebook allows users to combine programming with data visualization effectively.




````python
# Declaring variables
a = 15
b = 10

# Performing operations
sum_result = a + b
difference = a - b
product = a * b

# Displaying results
print("Sum:", sum_result)
print("Difference:", difference)
print("Product:", product)
````

###  Explanation of Code

In the above code:
- Two variables `a` and `b` are declared  
- Basic arithmetic operations like addition, subtraction, and multiplication are performed  
- The results are displayed using print statements  

This shows how Python can be used for simple calculations.

````python
import matplotlib.pyplot as plt

# Data for plotting
x = [1, 2, 3, 4, 5]
y = [3, 6, 9, 12, 15]

# Creating the plot
plt.plot(x, y, marker='o')

# Adding details
plt.title("Linear Growth Example")
plt.xlabel("X Values")
plt.ylabel("Y Values")

# Displaying grid
plt.grid()

# Show plot
plt.show()
````

### Graph Explanation

The above graph represents a linear relationship between X and Y values. As X increases, Y increases proportionally.

This type of visualization helps in understanding patterns and trends in data.
